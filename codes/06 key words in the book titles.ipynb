```python
import pandas as pd 
```

```python
df = pd.read_csv('cityu sociology.csv', encoding = 'utf8')
```

```python
df_titles = pd.DataFrame(df['titles'], columns = ["titles"])
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')+['!',',','.','?','-s','-ly','</s>','s',':','=']
df_titles['titles'] = df_titles['titles'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df_titles['titles'].head()
```

```python

freq_common1 = pd.Series(' '.join(df_titles['titles']).split())
text=list(freq_common1)


new_lists = []
for i in text:
    new_lists.append(i.lower())
print(new_lists)
```

```python
name=[]
number=[]
for each_n in set(new_lists):
    count = 0
    for each_Link in sorted(new_lists):
        if each_n == each_Link:
            count += 1
    print(each_n, ": ", count)
    name.append(each_n)
    number.append(count)
```

```python
import pandas as pd
dic = {
    'name': name,
    'number': number
}
   
pd_form = pd.DataFrame.from_dict(dic, orient = 'index').transpose()

#pd_form.to_csv('cuhkm1.csv')
#pd_form

pd_form.sort_values("number",inplace=True)

print(pd_form)
```

```python
file = open('cityusocial title.txt', 'w')
for t in new_lists:
    file.write(t)
    file.write('\n')
file.close()
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


filename = "cityusocial title.txt"
shape = "cityu.jpg"
font_path = "AdobeHebrew-Regular.otf"
cuhk_color = np.array(Image.open(shape))
image_colors = ImageColorGenerator(cuhk_color)



text = open(filename).read()

wd = WordCloud(
        width=1024,
        height=768,                  
        font_path=font_path,         
        background_color="white",    
        max_words=2000,             
        mask=cuhk_color,            
        max_font_size=250,          
        #min_font_size=10,
        random_state=42            
    )

wd.generate(text)

plt.figure()  

plt.imshow(wd.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")


plt.show()  
wd.to_file('cityusocial.png')   
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


filename = "cuhksocial title.txt"
#picture = "C:\\Users\\scott.wang\\Desktop\\test.png"
shape = "cuhk3.jpeg"
font_path = "AdobeHebrew-Regular.otf"
cuhk_color = np.array(Image.open(shape))
image_colors = ImageColorGenerator(cuhk_color)



text = open(filename).read()

wd = WordCloud(
        width=1024,
        height=768,                  # width,height设置生成的词云图片的大小
        font_path=font_path,         # 设置字体为本地的字体，有中文必须要加
        background_color="white",    # 设置背景的颜色，需与背景图片的颜色保持一致，否则词云的形状会有问题
        max_words=2000,              # 设置最大的字数
        mask=cuhk_color,             # 通过mask 参数 来设置背景图片，即词云的形状
        max_font_size=100,            # 设置字体的最大值
        #min_font_size=10,
        #stopwords=stopwords,         # 设置停用词
        random_state=42              # 设置有多少种随机生成状态，即有多少种配色方案
    )
# generate 可以对全部文本进行自动分词,但是他对中文支持不好，在WordCloud中设置字符的路径
wd.generate(text)


plt.figure()  # 生成一个新的图像
# 用词云形状的图片颜色来渲染词云的颜色，用color_func来指定
plt.imshow(wd.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")



plt.show()  # 展示所有的图片
wd.to_file('cuhksocial2.png')    #图片保存

```
