```python

import csv
Link=[]
with open('cuhk media.csv', 'r') as f:
    reader = csv.reader(f)
    for line in f:
        Link = [row[4] for row in reader]
        print(Link)
    
   
```

```python
sorted(Link)
```

```python
year=[]
cityu=[]
b = sorted(Link)
for each_b in b:
    count = 0
    for each_Link in sorted(Link):
        if each_b == each_Link:
            count += 1
    print(each_b, ": ", count)
    year.append(each_b)
    cityu.append(count)
    
```

```python
import pandas as pd
dic = {
    'year': year,
    'cityu': cityu
}
   
pd_form = pd.DataFrame.from_dict(dic, orient = 'index').transpose()

pd_form.to_csv('cuhkm1.csv')
pd_form
```
