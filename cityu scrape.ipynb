```python
import csv
Link=[]
with open('booklink.csv', 'r') as f:
    reader = csv.reader(f)
    for line in f:
        Link = [row[1] for row in reader]
        print(Link)
    
   
```

```python
from selenium import webdriver  
import time 
from lxml import etree
from selenium.common.exceptions import NoSuchElementException 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
import pandas as pd
driver = webdriver.Chrome()

titles = []
containers = []
authors = []
publishers = []
years = []
headers = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit/604.1.38 (KHTML, like Gecko) Version/11.0 Mobile/15A372 Safari/604.1'}
def read_html(url): # the URL is the webpage's url we want to scrape 
    print('start!')
    #i=0
    driver.get(url)      
    time.sleep(2)
    
    dtitles = driver.find_elements_by_class_name("item-title")
    for t in dtitles:
        print(t.text)
        titles.append(t.text)
        time.sleep(2) 
    
    auth = [pub.text for pub in driver.find_elements_by_class_name("item-detail")][0]
    pubs = [pub.text for pub in driver.find_elements_by_class_name("item-detail")][-1]
    print('auth',auth)
    print('pubs', pubs)
    publishers.append(pubs)
    authors.append(auth)
    time.sleep(2)

    # c2018
    year = pubs.split(' ')[-1].replace('c', '').replace('-', '')
    year = year if '1000' < year <= '2018' else ''
    print('year:', year)
    years.append(year)
    print('一本....\n')
    


        #i=i+1
time.sleep(2)
        
```

```python
for url in Link[:200]:
    time.sleep(15)
    print(url)
    read_html(url)
```

```python
dic = {
    'titles': titles,
    'publishers': publishers,
    'authors': authors,
    'years': years,
}
   
pd_form = pd.DataFrame.from_dict(dic, orient = 'index').transpose()

pd_form.to_csv('cityu media.csv')
#pd_form
```
