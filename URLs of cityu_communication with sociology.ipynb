Since we use the same code to collect the URLs of course reserve books in communication and sociology of cityu library, here we only use communication as an example.

```python
from selenium import webdriver  
import time 
import re
from lxml import etree
from bs4 import BeautifulSoup
import bs4 as bs
from selenium.common.exceptions import NoSuchElementException 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
import pandas as pd

#browser = webdriver.Firefox()#Chrome('./chromedriver.exe') 
LIBRARY_HOME_PAGE_URL = "https://julac.hosted.exlibrisgroup.com/primo-explore/search?query=sub,contains,communication,OR&query=sub,contains,media,AND&tab=default_tab&search_scope=CourseReserve&sortby=rank&vid=CUH&mode=advanced&offset=0" 
PATIENCE_TIME = 60 
LOAD_MORE_BUTTON_XPATH = '//*[@id="mainResults"]/div[2]/div/div[11]/button' 



driver = webdriver.Chrome() 
driver.get(LIBRARY_HOME_PAGE_URL) 
driver.maximize_window()
i=0
while True: 
    try: 
        loadMoreButton = driver.find_element_by_xpath("//button[contains(.,'Load more results')]") 
        time.sleep(10) 
        loadMoreButton.click() 
        print("success!")
        time.sleep(10)
        i = i+1
        if i==5:
            break
    except Exception as e: 
        print (e) 
        break 
print ("Complete")
time.sleep(5) 

```

```python
links = []
html = driver.page_source
soup = BeautifulSoup(html,'lxml')
linkcontainer = soup.find('div',class_='results-container zero-padding layout-column').find_all('div',class_='result-item-text layout-fill layout-column flex')
for container in linkcontainer:
        link = container.find('h3', class_='item-title').find(name='a',attrs={'class':'md-primoExplore-theme',"ng-href":re.compile(r'^https:')})
        links.append(link)
```

```python
print(links)
```

```python
Link=[]
for link in links:
    link = link.get('href')
    Link.append(link)
```

```python
print(Link)
```

```python
dic = {
    'Link': Link,
}
   
pd_form = pd.DataFrame.from_dict(dic, orient = 'index').transpose()

pd_form.to_csv('cityu_communication_booklink.csv')
```
